```python
import cv2
from google.colab.patches import cv2_imshow  # para evitar o problema com o Colab

# Carregar a imagem
img = cv2.imread('dog.png', cv2.IMREAD_GRAYSCALE)  # Carrega a imagem em escala de cinza

# Verificar se a imagem foi carregada corretamente
if img is None:
    print("Erro ao carregar a imagem.")
else:
    # Detectar bordas com o algoritmo Canny
    edges = cv2.Canny(img, 100, 200)  # Defina os limiares de detecção de bordas conforme necessário

    # Exibir a imagem original
    cv2_imshow(img)
```

```python
# Exibir as bordas detectadas
cv2_imshow(edges)
```
